# ECI, IECI and IECIR

```python
g=Graph.Erdos_Renyi(n=2000,m=3500)
```

```python
res=IECIR(g)
```

```python
labels=['ECI','IECI','IECIR']
for i,(x,y,_) in enumerate(res):
    plt.plot(x,y,label=labels[i])
plt.legend()
plt.xlabel('q')
plt.ylabel('GCC(q)')
plt.title('ER')
plt.show()
```
